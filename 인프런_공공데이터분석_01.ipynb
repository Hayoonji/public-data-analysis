### melt로 Tidy data만들기
pandas melt로 데이터형태 변형할수있음. df_first변수에 담긴 데이터프레임과 df_last에 담긴
데이터프레임 모습이 다르기에 같은 형태로 만들어줘야 데이터 합치기 가능.
데이터 병합위해 melt를 사용해 열에 있는 데이터를 행으로 녹여야됨

```python
#pd.melt사용하여 녹인데이터는 df_first_melt변수에 담기(df_last에 맞춰서 하는것)
df_first_melt=df_first.melt(id_vars='지역',var_name='기간',value_name='평당분양가격')
df_first_melt.head()
```

```python
#df_first_melt 변수에 담겨진 컬럼의 이름을
#['지역명','기간','평당분양가격']으로 변경(df_last 컬럼이름과 맞추기 위함)
df_first_melt.columns=['지역명','기간','평당분양가격']
df_first_melt.head(1)
```

### 연도와 월 분리하기 (df_last에선 따로 있어서)

```python
date='2013년12월'
date
```

```python
#split을 통해 '년'을 기준으로 텍스트 분리(리스트형태 반환)
date.split('년')
```

```python
#리스트 인덱싱을 사용해서 연도만 가져옴
date.split('년')[0]
```

```python
#리스트 인덱싱과 replace사용해서 월 제거
date.split('년')[1].replace('월','')
```

```python
#parse_year라는 함수를 만듦
# 연도만 반환하도록 하며, 반환하는 데이터는 int타입이 되도록함
def parse_year(date):
    year=date.split('년')[0]
    year=int(year)
    return year

y=parse_year(date)
y
```

```python
#제대로 분리가 되었는지 parse_year함수 확인
parse_year(date)
```

```python
#parse_month라는 함수를 만듦
#월만 반환하도록 하며, 반환하는 데이터는 int타입이 되도록함
def parse_month(date):
    month=date.split('년')[1].replace('월','')
    month=int(month)
    return month

m=parse_month(date)
m
```

```python
#제대로 분리가 되었는지 parse_month함수 확인
parse_month(date)
```

```python
#df_first_melt변수에 담긴 데이터프레임에서
#apply 활용해 연도만 추출해서 새로운 컬럼에 담기
df_first_melt['연도']=df_first_melt['기간'].apply(parse_year)
df_first_melt.head(1)
```

```python
#df_first_melt변수에 담긴 데이터프레임에서
#apply 활용해 월만 추출해서 새로운 컬럼에 담기
df_first_melt['월']=df_first_melt['기간'].apply(parse_month)
df_first_melt.head(1)
```

```python
#df_last와 병합위해서는 컬럼 이름 동일
#sample활용하여 데이터 미리보기
df_last.sample() #df_first와 df_last의 열 이름이 동일하기때문에 합쳐질 준비 완료
```
