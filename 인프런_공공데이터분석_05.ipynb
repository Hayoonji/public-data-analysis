### 편의시설

```python
df['공원보유시설(편익시설)'].value_counts()
```

```python
def get_string(text):
    if pd.isna(text):
        return text
    text=re.sub("[\d\.'\-(nan)\[\]\/]","",text)
    text=re.sub(" ","",text)
    text=re.sub(",{2,}",",",text)
        
    return text
```

```python
get_string("[1. '화장실' 5,2. 주차장, -, nan,/]")
```

```python
get_string(np.nan)
```

```python
df['편익시설']=df['공원보유시설(편익시설)'].map(get_string)
df['편익시설'].head()
```

```python
df['편익시설'].value_counts().tail()
```

```python
fac=','.join(df.loc[df['편익시설'].notnull(),'편익시설'])
fac=get_string(fac)
fac[:100]
```

```python
fac=str(df['편익시설'].tolist())
fac=get_string(fac)
fac[:100]
```

```python
#,가 2개이상 들어가면 ,를 1개로 바꾸는 정규표현식
re.sub(",{2,}",",", '화장실,주차장,,,')
```

```python
fac_count=pd.Series(fac.split(',')).value_counts()
fac_count.head()
```

```python
fac_count.head(50).plot.bar(figsize=(15,4))
```
